### Introduction

During this project, a lot of effort was put on gathering the data from different data sources to achieve the analysis deepth required on the WeRateDogs twitter account tweets.

### Gathering of data

WeRateDogs were kind enough to download and share with us their tweet archieve as they stood on August 1, 2017. The csv file twitter-archive-enhanced.csv

Also, the tweet images were passed through a prediciton alogirthm to determine if the images shared are for dogs and allow for filtering out non-dog tweets. This was programatically downloaded using the API and stored as image-predictions.tsv

Also, the tweeter API was queried for extra engagement metrics associated with the tweets to enrich the data and the analysis carried out. This is stored as tweet-json.txt.


### Inspection of Data

The download files were initially inspected visually using Excel for csv and tsv files, and also a text editor and https://jsonformatter.org/ for viewing different objects structure within the tweet-json.txt.

Once data was read into dataframes for csv and tsv files a programmatic assessment is carried out using df.info(), df.head and df.tail, and df.value_counts methods as well various other coding tenchiques to determine the extent of issues present in certain columns of interest.

A particular attention was placed on None, NA, empty strings, and attribute types. Also, value formatting and its sensibility was assessed for some columns.

The assessment revealed some dirty data present in our collated files due to the nature of the webs, human input inconsistencies, and also the accuracy of programmatic extraction.

In the end, a list of 8 quality issues and 2 tideness issues were identified to clean to make this data easily usable for further data wrangling and analysis for deriving insights on the tweeter account activity.

Generally speaking quality issues are concerned with the actual data present in columns whethe that is related to erroneous values, erroneous data types, and inconsistent formatting. Whereas tideness issues are more concerned with the structure of the data such as having a variable present in multiple columns, duplicate columns, fragmented data sets.

### Data Cleaning

Following the identification of those issues, a methodological process was followed in going through the issues one by one resolving those issues using the vast amount of python tools and packages available for us to use.

The process involved going through 3 steps for each issue:
- Define the issue in a more descriptive way
- Clean the data accordingly using Python code
- Test the new data correctness and suitability

The cleaned up dataframes, where then merged together to produce a master dataframe that has all the required metrics and dimensions needed for the data wrangling and analysis stage. Only records that are present in all of the three data sets were kept and other records missing in any of the data sets were thrown away.

### Save Data

The final clean dataframe is saved to a new csv file, so that it can be used anytime for further data wrangling and analysis without having to go through the cleaning steps above, leading to increased efficiences in later steps.
